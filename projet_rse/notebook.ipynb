```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/global_waste_data.csv')
df.head()
```

```
# Nettoyage
df = df.dropna()
df
```

```
# Top 10 pays produisant le plus de déchets
top_waste = df.nlargest(10, 'Total_Waste')
plt.figure(figsize=(10,6))
sns.barplot(data=top_waste, x='Country', y='Total_Waste')
plt.xticks(rotation=45)
plt.title('Top 10 – Production de déchets')
plt.show()
```

```
# Top 10 taux de recyclage
top_recycle = df.nlargest(10, 'Recycling_Rate')
plt.figure(figsize=(10,6))
sns.barplot(data=top_recycle, x='Country', y='Recycling_Rate')
plt.xticks(rotation=45)
plt.title('Top 10 – Taux de recyclage')
plt.show()
```

```
# Corrélation déchets vs recyclage
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='Total_Waste', y='Recycling_Rate')
plt.title('Relation entre déchets produits et recyclage')
plt.show()
```
